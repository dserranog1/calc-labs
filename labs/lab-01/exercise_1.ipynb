```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return x ** 2

x0 = 1
h = 0.0001

slope = (f(x0 + h) - f(x0)) / h

domain = np.linspace(0, 2, 100)
y = f(domain)
tangent = f(x0) + slope * (domain - x0)


plt.plot(domain, y, label="$f(x) = x^2$")
plt.plot(domain, tangent, "--", label="Recta tangente en $x=1$")
plt.scatter([x0], [f(x0)], color="red")
plt.title("Visualizacion de la derivada como pendiente")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def f(x):
  return np.sin(x)

x0 = np.pi/4
h = 0.0001

slope = (f(x0 + h) - f(x0)) / h

domain = np.linspace(0, np.pi, 100)
y = f(domain)
tangent = f(x0) + slope * (domain - x0)


plt.plot(domain, y, label="$f(x) =$ $\sin(x)$")
plt.plot(domain, tangent, "--", label="Recta tangente en $x=$$\pi$$/4 $")
plt.scatter([x0], [f(x0)], color="red")
plt.title("Visualizacion de la derivada como pendiente")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
